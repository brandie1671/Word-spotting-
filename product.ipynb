```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'nike_data_2022_09.csv'
df = pd.read_csv(file_path)

# Specify the product-related term to search for
product_term = 'soft fleece' 

# Count occurrences of the product term in each description
df['term_occurrences'] = df['description'].str.lower().str.count(product_term.lower())

# Visualize occurrences using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['term_occurrences'])
plt.xlabel('Product Index')
plt.ylabel('Occurrences of Term "{}"'.format(product_term))
plt.title('Occurrences of "{}" in Product Descriptions'.format(product_term))
plt.show()
```
